## 2\. Números

Preencha as células de código para preencher os valores de (A), (B) e (C) na tabela de ticket médio abaixo:

<br>


| Dia   | Valor Total Vendas | Qtd Total Vendas | Ticket Medio   |
|-------|--------------------|------------------|----------------|
| 19/01 | (A) = 961.56       | 3                | 320.52         |
| 20/01 | 834.47             | (B) = 7          | 119.21         |
| 23/01 | 15378.12           | 5                | (C) = 3075.62  |

```
# (A)

tkm_a = 320.52
qtd_va = 3
vtv_a = tkm_a * qtd_va
print(vtv_a)

tkm = vtv_a / qtd_va
print(tkm)
```

```
# (B)
vtv_b = 834.47
tkm_b = 119.21
qtv_b = vtv_b / tkm_b
print(qtv_b)
```

```
# (C)
vtv_c = 15378.12
qtd_vc = 5
tkm_c = vtv_c / qtd_vc
print(tkm_c)
```

## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'
cancao = cancao.upper()
print(cancao)

cancao = cancao.replace('RODA', "porta")
print(cancao)

cancao = cancao.capitalize()
print(cancao)
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

busca_selic = noticia.find('2,75%')
print(busca_selic)

busca_selic = noticia[12:17]
print(busca_selic)

busca_ano = noticia.find('6')
print(busca_ano)
busca_ano = noticia[62]
print(busca_ano)
print(len(noticia))
```
